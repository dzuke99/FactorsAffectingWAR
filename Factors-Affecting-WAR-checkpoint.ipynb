```python
def scrape_team(team,season,team_url_postfix):
    url = "https://www.fangraphs.com" + team_url_postfix + "/stats?season=" + str(season)
    print(url)
    html = urlopen(url)
    bs = BeautifulSoup(html.read(), 'lxml')
    tables = bs.find_all('table',{"class":"tablesort"})
    stats_tables = pd.read_html(str(tables))
    
    batting = stats_tables[0][:-1]
    batting['Team'] = team
    batting['Season'] = season
    
    starting_pitching = stats_tables[1][:-1]
    starting_pitching['Team'] = team
    starting_pitching['Season'] = season
    
    relief_pitching = stats_tables[2][:-1]
    relief_pitching['Team'] = team
    relief_pitching['Season'] = season
    
    fielding = stats_tables[3][:-1]
    fielding['Team'] = team
    fielding['Season'] = season
    return(batting,starting_pitching,relief_pitching,fielding)

batting, starting_pitching, relief_pitching, fielding = scrape_team("Braves",2019,"/teams/braves") 
```

```python
html = urlopen('https://www.fangraphs.com/teams/braves/stats?season=2019')
bs = BeautifulSoup(html.read(), 'lxml')
teams = bs.find("select",{"class":"select-change-team"}).find_all("option")
```

```python
teams_batting = pd.DataFrame()
teams_starting_pitching = pd.DataFrame()
teams_relief_pitching = pd.DataFrame()
teams_fielding = pd.DataFrame()
seasons = np.arange(2017,2020)
for season in seasons:
    html = urlopen('https://www.fangraphs.com/teams/mets/stats?season=' + str(season))
    bs = BeautifulSoup(html.read(), 'lxml')
    teams = bs.find("select",{"class":"select-change-team"}).find_all("option")
    for team in teams:
        team_url_postfix = team.attrs.get('value')
        if(team_url_postfix == None):
            continue
        team = team.attrs.get('label')
        team_batting,team_starting_pitching,team_relief_pitching,team_fielding = \
            scrape_team(team,season,team_url_postfix)
        teams_batting = pd.concat([teams_batting,team_batting])
        teams_starting_pitching = pd.concat([teams_starting_pitching,team_starting_pitching])
        teams_relief_pitching = pd.concat([teams_relief_pitching,team_relief_pitching])
        teams_fielding = pd.concat([teams_fielding,team_fielding])
        time.sleep(np.random.uniform(2,5))
```

```python
teams_batting.to_csv("./teams_batting.csv")
teams_starting_pitching.to_csv("./teams_starting_pitching.csv")
teams_relief_pitching.to_csv("./teams_relief_pitching.csv")
teams_fielding.to_csv("./teams_fielding.csv")
```
